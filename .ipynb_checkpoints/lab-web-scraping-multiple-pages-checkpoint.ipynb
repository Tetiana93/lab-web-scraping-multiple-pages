```python
artists=[]
for i in range(1, 101):
    r=requests.get("https://pitchfork.com/reviews/albums/?page=1")
    html = BeautifulSoup(r.content, 'html.parser')
    artists.extend([(s.get_text(strip=True)) for s in html.find_all('ul', attrs={'class':'artist-list review__title-artist'})])
```

```python
artists
```

```python
songs=[]
for i in range(1, 101):
    r=requests.get("https://pitchfork.com/reviews/albums/?page=1")
    html = BeautifulSoup(r.content, 'html.parser')
    songs.extend([(s.get_text(strip=True))for s in html.find_all('h2', attrs={'class':'review__title-album'})])
```

```python
[(s.get_text(strip=True))for s in html.find_all('h2', attrs={'class':'review__title-album'})]
```

```python
songs
```

```python
df=pd.DataFrame({'Column1': songs, 'Column2': artists})
df
```

```python
df=df.rename(columns={'Column1': 'Song', 'Column2': 'Artist'})
df
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configure Selenium to use a headless browser
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(options=options)

# Navigate to the page
driver.get('https://example.com')

# Wait for the dynamic element with a specific href to appear
dynamic_element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, 'a[href="//www.emsc-csem.org/Earthquake_information/"]'))
)

# Extract the text or other attributes of the dynamic element if needed
dynamic_element_text = dynamic_element.text
dynamic_element_href = dynamic_element.get_attribute('href')

# Process or store the scraped content as needed
print(dynamic_element_text)
print(dynamic_element_href)

# Close the browser
driver.quit()

```

```python
r=requests.get('https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers')
r.status_code
```

```python
soup=BeautifulSoup(r.content, 'html.parser')
soup_p=soup.prettify()
print(soup_p)
```

```python
language=[(s.get_text(strip=True)) for s in soup.find_all('a', attrs={'class':'mw-redirect'})]
language
```

```python
len(language)
```

```python
del language[:3]
language
```

```python
language=language[:10]
language
```

```python
native_speakers=[(s.get_text(strip=True)) for s in soup.find_all('td') if s.get_text(strip=True).isnumeric()]
```

```python
native_speakers=native_speakers[:10]
native_speakers
```

```python
df=pd.DataFrame({'language':language,'native_speakers':native_speakers})
df
```
